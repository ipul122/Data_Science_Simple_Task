```python
import pandas as pd

df = pd.read_csv("../Pandas_DF/data.csv")
print(df)
```

## Data Index Filtering

```python
df.iloc[0]["City"]   ## Single value
```

```python
df.iloc[0:3]        ## Multiple rows range value
```

```python
df.iloc[0:3][["City","Value"]]        ## Multiple rows multiple columns
```

# Filtering Using Criteria

```python
df
```

```python
df[(df["City"] == "Lampung")]        
```

```python
df[(df["Value"] <= 5)]
```

```python
df[~(df["Value"] <= 5)]
```

```python
df[(df["Value"] >= 5) & (df["City"] == "Surabaya")]
```

```python
list_city = ["Bogor", "Yogyakarta", "Semarang", "Surabaya", "Bandung", "Medan", "Makassar", "Palembang", "Denpasar", "Balikpapan"]

df[
    (df["City"].isin(list_city)) &
    (df["Value"] >= 5)
]
```
