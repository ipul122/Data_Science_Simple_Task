```python
import pandas as pd 
import numpy as np 
```

# List

```python
list_of_data =[["C1", "Lampung", 3], ["C2", "Jakarta", 6], ["C3", "Bogor", 9 ], ["C4", "Tanggerang", 9 ], ["C5", "Cikarang", 7 ]]
```

```python
df = pd.DataFrame(list_of_data, columns=["ID", "City", "Value"])
print(df)
```

# Array

```python
numpy_array = np.array(list_of_data)
print(numpy_array)
```

```python
df_numpy = pd.DataFrame(numpy_array, columns=["ID", "City", "Value"])
print(df_numpy)
```

# Dict

```python
dict_of_data = {"ID": ["C1", "C2", "C3", "C4", "C5"], 
                "City": ["Lampung", "Jakarta", "Bogor", "Tanggerang", "Cikarang"], 
                "Value": [3, 6, 9, 9, 7]}
print(dict_of_data)
```

```python
df_dict = pd.DataFrame(dict_of_data)
print(df_dict)
```
